```python
import numpy as np
import matplotlib.pyplot as plt

pathplot = '/home/lukas/Schreibtisch/SModelS/code/Higgs/output/plots/'
```

```python
path_true = 'output/True_para.out'
path_false = 'output/False_para.out'

f = open(path_true, 'r')
list_true = f.readlines()
f.close
f = open(path_false, 'r')
list_false = f.readlines()
f.close

print(len(list_true))
print(len(list_false))
```

```python
para_true = []
para_false = []

para_name = list_true[0].split('\t')[:-1]
for element in list_true[1:]:
    item = np.array(str(element).split('\t'))[:-1]
    if para_name[0] == item[0]: continue
    item = [float(par) for par in item]
    para_true.append(np.asarray(item))

for element in list_false[1:]:
    item = np.array(str(element).split('\t'))[:-1]
    if para_name[0] == item[0]: continue
    item = [float(par) for par in item]
    para_false.append(np.asarray(item))

para_min = [1,500,1200,1,100,5,250,500,500]
para_max = [500,1500,2500,5000,500,60,1000,1500,1500]
slices = [2,3,3,4,2,4,2,3,3]
# slices = [2,2,2,2,2,2,2,2,2]
unit = ['GeV','GeV','GeV','GeV','GeV','','GeV','GeV','GeV']
para_name_latex = ['M_1','M_2','M_3','A_t','\mu','tan b','M_A','M_{Q_3}','M_{UR_3}']

print(len(para_true))
print(len(para_false))

fontsize = 12
legfontsize = 10
```

```python
# # para_name
# # para_true[-1]
# # range(len(para_true[0])-1)
# # np.matrix(para_true).T[0]

# i = 1
# j = 4
# min = []
# max = []
# for k in range(9):
#     if k == i or k==j: continue
#     min.append([int(para_min[k] + (para_max[k]-para_min[k])*slice/slices[k]) for slice in range(slices[k])])
#     max.append([int(para_min[k] + (para_max[k]-para_min[k])*slice/slices[k]) for slice in range(1,slices[k]+1)])
#     valt = np.matrix(para_true).T[k]
#     valf = np.matrix(para_false).T[k]

# h=0
# for a in range(len(min[0])):
#     for b in range(len(min[1])):
#         for c in range(len(min[2])):
#             for d in range(len(min[3])):
#                 for e in range(len(min[4])):
#                     for f in range(len(min[5])):
#                         for g in range(len(min[6])):
#                             para_list_min = [min[0][a],min[1][b],min[2][c],min[3][d],min[4][e],min[5][f],min[6][g]]
#                             para_list_max = [max[0][a],max[1][b],max[2][c],max[3][d],max[4][e],max[5][f],max[6][g]]
#                             print(para_list_min)
#                             print(para_list_max)
#                             h = h+1
# print(h)                         
```

```python
for i in range(len(para_true[0])-1):
    for j in range(len(para_true[0])-1):
#        if i == j: continue
        if i >= j: continue

        min = []
        max = []
        for k in range(len(para_true[0])-1):
            if k == i or k==j: continue
#                 min.append([float('nan') for slice in range(slices[k])])
#                 max.append([float('nan') for slice in range(1,slices[k]+1)])
            else:
                min.append([int(para_min[k] + (para_max[k]-para_min[k])*slice/slices[k]) for slice in range(slices[k])])
                max.append([int(para_min[k] + (para_max[k]-para_min[k])*slice/slices[k]) for slice in range(1,slices[k]+1)])

        num = 0
        for a in range(len(min[0])):
            for b in range(len(min[1])):
                for c in range(len(min[2])):
                    for d in range(len(min[3])):
                        for e in range(len(min[4])):
                            for f in range(len(min[5])):
                                for g in range(len(min[6])):
                                    num += 1
                                    para_list_min = [min[0][a],min[1][b],min[2][c],min[3][d],min[4][e],min[5][f],min[6][g]]
                                    para_list_max = [max[0][a],max[1][b],max[2][c],max[3][d],max[4][e],max[5][f],max[6][g]]
                                    
                                    if i<j:
                                        para_list_min.insert(i, float('nan'))
                                        para_list_min.insert(j, float('nan'))
                                        para_list_max.insert(i, float('nan'))
                                        para_list_max.insert(j, float('nan'))
                                    else:
                                        para_list_min.insert(j, float('nan'))
                                        para_list_min.insert(i, float('nan'))
                                        para_list_max.insert(j, float('nan'))
                                        para_list_max.insert(i, float('nan'))

                                    xt = np.array(np.matrix(para_true).T[i])[0]
                                    yt = np.array(np.matrix(para_true).T[j])[0]
                                    xf = np.array(np.matrix(para_false).T[i])[0]
                                    yf = np.array(np.matrix(para_false).T[j])[0]

                                    for k in range(len(para_true[0])-1):
                                        if k == i or k==j: continue
                                        valt = np.array(np.matrix(para_true).T[k])[0]
                                        valf = np.array(np.matrix(para_false).T[k])[0]
                                        top = int(para_list_max[k])
                                        bottom = int(para_list_min[k])
                                        xt = [xt[l] if valt[l] < top and valt[l] > bottom else float('nan') for l in range(len(xt))]
                                        yt = [yt[l] if valt[l] < top and valt[l] > bottom else float('nan') for l in range(len(yt))]
                                        xf = [xf[l] if valf[l] < top and valf[l] > bottom else float('nan') for l in range(len(xf))]
                                        yf = [yf[l] if valf[l] < top and valf[l] > bottom else float('nan') for l in range(len(yf))]

                                    xt = [item for item in xt if item == item]
                                    yt = [item for item in yt if item == item]
                                    xf = [item for item in xf if item == item]
                                    yf = [item for item in yf if item == item]

                                    if len(xt) == 0 and len(xf) == 0:
                                        continue
                                    labelx = para_name_latex[i]
                                    labely = para_name_latex[j]


                                    plt.figure()
                            #         if i in [0,1,2,3,4,6,7,8]: plt.semilogx()
                            #         if j in [0,1,2,3,4,6,7,8]: plt.semilogy()
                                    axes = plt.gca()
                                    axes.set_xlim([para_min[i],para_max[i]*1.3])
                                    axes.set_ylim([para_min[j],para_max[j]])

                                    plt.rcParams.update({'font.size': fontsize})

                                    plt.scatter(xf, yf, marker='.', color='red', label=r'excluded')
                                    plt.scatter(xt, yt, marker='.', color='green', label=r'not excluded')
                                    plt.scatter([0,0], [0,0], color='green', label='HiggsBounds-4.3.1')
                                    plt.scatter([0,0], [0,0], color='green', label='HiggsSignals-1.4.0')
                                    for k in range(len(para_true[0])-1):
                                        if k == i or k==j: continue
                                        plt.scatter([0,0], [0,0], color='green', label=r'$ ' + str(para_list_min[k]) + ' < ' + para_name_latex[k] + ' < ' + str(para_list_max[k]) + ' $ [' + unit[k] + ']')

                                    leg = plt.legend(loc='upper right', fancybox=True, framealpha=1., fontsize=legfontsize)
                                    for item in leg.legendHandles[2:]:
                                        item.set_visible(False)
                                    axes.set_title('parameter exclusion, $p < 0.05$')
                                    plt.xlabel(r'$' + labelx + '$ [' + unit[i] + ']')
                                    plt.ylabel(r'$' + labely + '$ [' + unit[j] + ']')
                                    plt.savefig(pathplot + 'exclusion_' + para_name[i] + '_' + para_name[j] + '_p005_' + str(num).zfill(4) + '.pdf', format='pdf')
                                    plt.savefig(pathplot + 'exclusion_' + para_name[i] + '_' + para_name[j] + '_p005_' + str(num).zfill(4) + '.png', format='png')
                            #         plt.show()
                                    plt.cla()
                                    plt.clf()
                                    plt.close('all')
```
